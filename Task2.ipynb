2.Cleaning Data

```python
df.isnull().sum()
```

```python
numericcolumns = ["Sales", "Profit", "Quantity", "Discount"]
for col in numericcolumns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print(df.select_dtypes(include='number').dtypes)
```

```python
df=df[df["Sales"] >= 0]
df=df[df["Quantity"] >= 0]
```

```python
df = df[(df["Discount"] >= 0) & (df["Discount"] <= 1)]
```

3.1 Feature Extraction

```python
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
```

```python
df["Year"]=df["Ship Date"].dt.year
```

```python
df["Month"]=df["Ship Date"].dt.month
```

```python
df.columns
```

```python
df['Unit_Price'] = np.random.uniform(10, 100, len(df))
```

```python
df
```

```python
df['Total_Revenue']= df['Quantity']*df['Unit_Price']
print(df)
```

```python
df['Unit_cost']= np.random.uniform(10, 100, len(df))
```

```python
df["Total Cost"]= df["Quantity"]*df["Unit_cost"]
print(df)
```

```python
df["Profit"]= df["Total_Revenue"]- df["Total Cost"]
print(df)
```

```python
def clas_profit(x):
    if x > 0:
        return"Positive"
    elif x < 0:
        return"Negative"
    else:
        return"Break Even"
df["ProfitStatus"] = df["Profit"].apply(clas_profit)
print(df)
```
